```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Example data (replace with your own data)
social_media_data = [
    "I love this brand! Their products are amazing.",
    "I had a terrible experience with their customer service.",
    "Neutral tweet about the topic."
    # Add more social media posts here
]

# Preprocessing function
def preprocess_text(text):
    # Add your preprocessing steps here (cleaning, tokenization, etc.)
    return text

# Perform sentiment analysis
sentiments = []
for post in social_media_data:
    cleaned_post = preprocess_text(post)
    analysis = TextBlob(cleaned_post)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        sentiments.append('Positive')
    elif polarity < 0:
        sentiments.append('Negative')
    else:
        sentiments.append('Neutral')

# Visualize sentiment distribution
sentiment_counts = pd.Series(sentiments).value_counts()
plt.figure(figsize=(6, 6))
sentiment_counts.plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red', 'gray'])
plt.title('Sentiment Distribution')
plt.axis('equal')
plt.show()
```
